```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('./opencv_logo.PNG', cv2.IMREAD_COLOR)
```

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
bin_img = cv2.threshold(gray_img, 250, 255, cv2.THRESH_BINARY)[1]
```

```python
fig, subs = plt.subplots(ncols = 3, sharex = True, sharey = True, figsize = (15, 5))
subs[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
subs[1].imshow(gray_img, cmap = 'gray')
subs[2].imshow(bin_img, cmap = 'gray')
plt.show()
```

```python
grad = np.zeros((256, 256))
for i in range(256):
    for j in range(256):
        grad[i, j] = j
```

```python
fig, subs = plt.subplots(ncols = 3, nrows = 2, sharex = True, sharey = True, figsize = (9, 6))

subs[0, 0].set_title('Origin')
subs[0, 0].imshow(grad, cmap = 'gray')

subs[0, 1].set_title('THRESH_BINARY')
subs[0, 1].imshow(cv2.threshold(grad, 127, 255, cv2.THRESH_BINARY)[1], cmap = 'gray')

subs[0, 2].set_title('THRESH_BINARY_INV')
subs[0, 2].imshow(cv2.threshold(grad, 127, 255, cv2.THRESH_BINARY_INV)[1], cmap = 'gray')

subs[1, 0].set_title('THRESH_TRUNC')
subs[1, 0].imshow(cv2.threshold(grad, 127, 255, cv2.THRESH_TRUNC)[1], cmap = 'gray')

subs[1, 1].set_title('THRESH_TOZERO')
subs[1, 1].imshow(cv2.threshold(grad, 127, 255, cv2.THRESH_TOZERO)[1], cmap = 'gray')

subs[1, 2].set_title('THRESH_TOZERO_INV')
subs[1, 2].imshow(cv2.threshold(grad, 127, 255, cv2.THRESH_TOZERO_INV)[1], cmap = 'gray')

plt.show()
```

```python
sudoku = cv2.imread('./sudoku.jpg', cv2.IMREAD_GRAYSCALE)

plt.imshow(sudoku, cmap = 'gray')
plt.show()
```

```python
bin50  = cv2.threshold(sudoku,  50, 255, cv2.THRESH_BINARY)[1]
bin100 = cv2.threshold(sudoku, 100, 255, cv2.THRESH_BINARY)[1]
bin150 = cv2.threshold(sudoku, 150, 255, cv2.THRESH_BINARY)[1]

fig, subs = plt.subplots(ncols = 3, figsize = (15, 5), sharex = True, sharey = True)

subs[0].set_title('thresh : 50')
subs[0].imshow(bin50, cmap = 'gray')

subs[1].set_title('thresh : 100')
subs[1].imshow(bin100, cmap = 'gray')

subs[2].set_title('thresh : 150')
subs[2].imshow(bin150, cmap = 'gray')

plt.show()
```
